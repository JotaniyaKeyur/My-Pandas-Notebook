drop_duplicate - used to remove duplicated

```
#This will remove all duplicates in city means all city in data will be unique
df.drop_duplicates(subset=["city"])
```

```
#It also provides which occurence to keep first or last
#It have 3 args(subset,keep,inplace) if inplace is true changes will occur in original data.
df.drop_duplicates(subset=['city']).shape
```

```
#We can also apply it to multiple columns
df.drop_duplicates(subset=["city", "season"])
```

```
df.drop_duplicates(subset=["city", "season"]).shape
```

```
#now suppose we need to find winner of each season of ipl
#So here we dropped duplicates from season and kept the last occurence so the last match will be the final match and we only
#wanted two columns and we sorted the data with season wise to get ascending order.

df.drop_duplicates(subset=['season'],keep='last')[['season','winner']].sort_values('season',ascending=True)
```

Handling Missing Values

dropna()

```
data = pd.read_csv("train.csv")
data
```

```
data.dropna()
#we can see rows are deleted which had NAN values
```

```
#by default here axis=0 which mean it will drop rows but if we do axis =1 it will drop columns
data.dropna(axis=1)
```

fillna()

```
data.fillna(0)
#Its not good as we are also filling 0 in string column
```

```
#so instead of applying fillna to entire dataframe we can instead apply it on columns, for permanent changes use inplace=True
data['Cabin'].fillna("Not Specified")
```

```
data['Age'].fillna(method='bfill')
#here bfill fills the value from previous value
#we can use ffill for forward value
```
